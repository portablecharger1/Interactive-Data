```python
import plotly.express as px
import plotly as pl
import pandas as pd
```

```python
food = pd.read_csv('daily_food_delivery_orders.csv')
food.head()
```

```python
status_counts= food["restaurant_type"].value_counts().reset_index()
status_counts.columns = ['restaurant_type', 'frequency'] 

fig1 = px.bar(status_counts, x='restaurant_type', y= 'frequency', title= 'frequency of Restaurant Type', labels={'restaurant_type': 'restaurant type', 'frequency': 'count'}, 
              text= 'frequency', color= 'restaurant_type', color_discrete_sequence=['brown', "#421B04", 'red', 'yellow', 'green', 'orange'])
fig1.show()
```

```python
food20= food.head(20)
food20
```

```python
fig2 = px.line(food20, x='order_date', y='order_value', markers=True)
fig2.show()
```

```python
food20['in_order_dates'] = pd.to_datetime(food['order_date'])
food20.sort_values('in_order_dates', inplace=True)
food20
```

```python
fig3 = px.line(food20, x='in_order_dates', y='order_value', markers=True, labels=dict(in_order_dates = 'Order Dates', order_value = 'Order Value'))

fig3.update_traces(marker=dict(color='red', size= 10, symbol= 'square'), line=dict(width=3, color='pink'))
fig3.show()
```

```python
fig4 = px.scatter(food, x= 'delivery_distance_km', y='delivery_time_minutes', color= 'delivery_partner_rating', symbol= 'order_status')

fig4.show()
```

```python
foodSamp = food.sample(n=50)
foodSamp.head()
```

```python
fig5 = px.scatter(foodSamp, x= 'delivery_distance_km', y='delivery_time_minutes', color= 'delivery_partner_rating', symbol= 'order_status', 
title= 'Delivery Distince vs. Delivery time', 
labels= {'delivery_distance_km' : 'Delivery Distance Km', 'delivery_time_minutes' : 'Delivery Time (min)', 
         'delivery_partner_rating' : 'Delivery Partner Rating', 'order_status' : 'Order Status'})
fig5.update_traces(marker= dict(size = 20, opacity = 0.5, line= dict(width = 2, color='black')))
fig5.update_layout(legend= {'yanchor': 'top', 'y': 0.88, 'xanchor': 'left', 'x':1.1})

fig5.show()
```

```python
plot_template = 'plotly_white'

status_color = {'Delivered' : "#2A9D4F", 'Delayed' : "#F9E833", 'Cancelled' : "#DB1021"}

food['age_group'] = pd.cut(food['customer_age'], bins=[0, 18, 25, 35, 45, 60, 100], labels= ['<18', '18-25', '26-35', '36-45', '46-60', '60+'])
```

```python
fig7 = px.histogram(food, x='order_value', color= 'order_status', color_discrete_map= status_color, facet_col='age_group', facet_col_wrap=3, nbins=30, 
                    title='Order Value Distrabution by Age and Status', labels={'order_value': 'Order Value'})

fig7.update_layout(template=plot_template)
fig7.show()
```
