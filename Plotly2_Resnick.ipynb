```python
pokedata= pd.DataFrame(pokedex['pokemon'])
pokedata.head(10)
```

```python
pokedata[['type1', 'type2']] = pd.DataFrame(pokedata.type.tolist(), index=pokedata.index)
pokedata.head()
```

```python
typegroup=pokedata.groupby('type1').count()['id']
typegroup
```

```python
pokecolors_map = {
'Normal': '#A8A77A',
'Fire': '#EE8130',
'Water': '#6390F0',
'Electric': '#F7D02C',
'Grass': '#7AC74C',
'Ice': '#96D9D6',
'Fighting': '#C22E28',
'Poison': '#A33EA1',
'Ground': '#E2BF65',
'Flying': '#A98FF3',
'Psychic': '#F95587',
'Bug': '#A6B91A',
'Rock': '#B6A136',
'Ghost': '#735797',
'Dragon': '#6F35FC',
'Dark': '#705746',
'Steel': '#B7B7CE',
'Fairy': '#D685AD',
}
```

```python
pokecolors=typegroup.index.map(pokecolors_map)
pokecolors=list(pokecolors)
pokecolors
```

```python
fig1=go.Figure(data=[go.Bar(x=typegroup.index, y=typegroup.values)])
fig1.show()
```

```python
fig1=go.Figure(data=[go.Bar(x=typegroup.index, y=typegroup.values, marker_color= pokecolors)])
fig1.show()
```

```python
fig2 = go.Figure(data=[go.Pie(labels= typegroup.index, values=typegroup.values, marker= dict(colors=pokecolors), title= 'Gen 1 Pokemon by First Type')])
fig2.show()
```

```python
conditions = [(pokedata["next_evolution"].apply(isinstance, args=(list,))) & (pokedata["prev_evolution"].apply(isinstance, args=(float,))),
              (pokedata["next_evolution"].apply(isinstance, args=(float,))) & (pokedata["prev_evolution"].apply(isinstance, args=(list,))),
              (pokedata["next_evolution"].apply(isinstance, args=(float,))) & (pokedata["prev_evolution"].apply(isinstance, args=(float,)))
              ]
choices = ["First", "Final", "Lone"]



pokedata['evolution_stage'] = np.select(conditions, choices, default="Middle")
pokedata.tail()
```

```python
evolvegroup=pokedata.groupby('evolution_stage').count()['id']
evolvegroup
```

```python
fig3= make_subplots(rows=1, cols=2, specs=[[{'type': 'domain'}, {'type': 'domain'}]])

fig3.add_trace(go.Pie(labels= typegroup.index, values=typegroup.values, marker= dict(colors=pokecolors), title= 'Gen 1 Pokemon by First Type'), 1,1)

fig3.add_trace(
    go.Pie(labels=evolvegroup.index, values=evolvegroup.values, title= 'Gen 1 Pokemon by Evolution Stage'), 1,2
)
fig3.show()
```

```python
pokedata['weight_kg'] = pokedata['weight'].str.replace(' kg', '', regex=False)
pokedata['weight_kg'] = pd.to_numeric(pokedata['weight_kg'])
```

```python
pokedata['height_m'] = pokedata['height'].str.replace(' m', '', regex=False)
pokedata['height_m'] = pd.to_numeric(pokedata['height_m'])

pokedata.head()
```

```python
traces = go.Scatter(
    x = pokedata['height_m'],
    y = pokedata['weight_kg'],
    mode = 'markers'
)

layout = go.Layout(title = {'text': 'Pokemon Weight and Height'},
                   xaxis = {'title': 'Weight in kg'},
                   yaxis = {'title': 'Height in m'})

fig4 = go.Figure(data = [traces], layout = layout)

fig4.show()
```

```python
condition = pokedata['type1'].notna() & pokedata['type2'].notna()
pokedata['types'] = np.where(condition, pokedata['type1'] + ' & ' + pokedata['type2'], pokedata['type1'])
pokedata.head()
```

```python
trace1 = go.Scatter(
    x = pokedata[pokedata['evolution_stage'] == 'Lone']['height_m'],
    y = pokedata[pokedata['evolution_stage'] == 'Lone']['weight_kg'],
    customdata = pokedata[pokedata['evolution_stage'] == 'Lone']['name'],
    name = 'Lone'
)

trace2 = go.Scatter(
    x = pokedata[pokedata['evolution_stage'] == 'First']['height_m'],
    y = pokedata[pokedata['evolution_stage'] == 'First']['weight_kg'],
    customdata = pokedata[pokedata['evolution_stage'] == 'First']['name'],
    name = 'First'
)

trace3 = go.Scatter(
    x = pokedata[pokedata['evolution_stage'] == 'Middle']['height_m'],
    y = pokedata[pokedata['evolution_stage'] == 'Middle']['weight_kg'],
    customdata = pokedata[pokedata['evolution_stage'] == 'Middle']['name'],
    name = 'Middle'
)

trace4 = go.Scatter(
    x = pokedata[pokedata['evolution_stage'] == 'Final']['height_m'],
    y = pokedata[pokedata['evolution_stage'] == 'Final']['weight_kg'],
    customdata = pokedata[pokedata['evolution_stage'] == 'Final']['name'],
    name = 'Final'
)

layout = go.Layout(title = {'text': 'Pokemon Weight and Height'},
                   xaxis = {'title': 'Weight in kg'},
                   yaxis = {'title': 'Height in m'})

fig4 = go.Figure(data = [trace1, trace2, trace3, trace4], layout = layout)

fig4.update_traces(mode = 'markers')

fig4.show()
```
